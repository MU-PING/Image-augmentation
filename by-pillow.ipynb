```python
import sys
import tensorflow as tf
import matplotlib.pyplot as plt

from PIL import Image 
from tensorflow.keras.datasets import cifar10
```

```python
print("Environment confirmation:", sys.executable)
print("TensorFlow version:", tf.__version__)
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# image upsampling
title = ["NEAREST", "BOX", "BILINEAR", "HAMMING", "BICUBIC", "LANCZOS"]
resampling = [Image.NEAREST, Image.BOX, Image.BILINEAR, Image.HAMMING, Image.BICUBIC, Image.LANCZOS]
image = x_train[0]

plt.figure(figsize=(20, 6))
ax = plt.subplot(1, 10, 1)
ax.imshow(image)
ax.set_title("ORIGIN", fontsize=14)

for i in range(6):
    ax = plt.subplot(1, 7, i+2)
    img = Image.fromarray(image)
    im_resize0 = img.resize((224, 224), resampling[i])
    ax.imshow(im_resize0)
    ax.set_title(title[i], fontsize=14)
```
