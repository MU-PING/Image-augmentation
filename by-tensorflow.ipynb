```python
import sys
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
```

```python
print("Environment confirmation:", sys.executable)
print("TensorFlow version:", tf.__version__)

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
image = x_train[0]
```

```python
# upsampling and crop and flip horizontal
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.Resizing(40, 40, interpolation='bilinear'),
    tf.keras.layers.RandomCrop(32, 32),
    tf.keras.layers.RandomFlip('horizontal')
])

# trainig
plt.figure(figsize=(20, 6))
ax = plt.subplot(1, 10, 1)
ax.imshow(image)
ax.set_title("ORIGIN", fontsize=14)

for i in range(6):
    augmented_image = data_augmentation(image, training=True)/255
    ax = plt.subplot(1, 7, i+2)
    ax.set_title("AUGMENTED", fontsize=14)
    plt.imshow(augmented_image)
```

```python
# inference test Resizing - still work
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.Resizing(1000, 1000, interpolation='bilinear'),
])

# trainig
plt.figure(figsize=(20, 6))
ax = plt.subplot(1, 10, 1)
ax.imshow(image)
ax.set_title("ORIGIN", fontsize=14)

for i in range(6):
    augmented_image = data_augmentation(image, training=False)/255
    ax = plt.subplot(1, 7, i+2)
    ax.set_title("AUGMENTED", fontsize=14)
    plt.imshow(augmented_image)
```

```python
# inference test RandomCrop - still work but crop center
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomCrop(16, 16),
])

# trainig
plt.figure(figsize=(20, 6))
ax = plt.subplot(1, 10, 1)
ax.imshow(image)
ax.set_title("ORIGIN", fontsize=14)

for i in range(6):
    augmented_image = data_augmentation(image, training=False)/255
    ax = plt.subplot(1, 7, i+2)
    ax.set_title("AUGMENTED", fontsize=14)
    plt.imshow(augmented_image)
```
